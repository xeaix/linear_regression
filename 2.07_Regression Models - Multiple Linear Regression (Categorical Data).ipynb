```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv('MLR_mix.csv')
ds.head()
```

```python
ds.tail()
```

```python
ds.info()
```

```python
corr = ds.corr()
print(corr)
```

```python
x = ds.iloc[:, [0, 1, 2, 3]].values
y = ds.iloc[:, 4].values
print(x)
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 3] = le.fit_transform(x[:, 3])
x
```

```python
le.inverse_transform([0, 1])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('State', OneHotEncoder(), [3])], remainder = 'passthrough')
x = ct.fit_transform(x)
x
```

```python
x = x[:, 1:] #keep B
#x = x[:, [0, 2, 3, 4]] #keep A
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
print('x_train:', x_train)
print('x_test:', x_test)
print('y_train:', y_train)
print('y_test:', y_test)
```

```python
# y = mx + c
# y = Bx + a
# m = B = gradient = coefficient
# c = a = y-intercept = constant

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
print('Coefficients, beta:', regressor.coef_)
print('Constant, alpha:', regressor.predict([[0, 0, 0, 0]]))
```

```python
# equal to 1 is good
# equal to 0 is not good

x = regressor.score(x_train, y_train)
print(x)
```

```python
y_pred = regressor.predict(x_test)
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import r2_score
y = r2_score(y_pred, y_test)
print(y)
```

```python
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
print(MSE)
```

```python
# Estimate the profit gained if:
# S&P department spend=90020,
# Admin department spend = 55200,
# Operation department spend = 94230
# State = B

y_pred = regressor.predict([[1, 90020, 55200, 94230]])
print(y_pred)
```
